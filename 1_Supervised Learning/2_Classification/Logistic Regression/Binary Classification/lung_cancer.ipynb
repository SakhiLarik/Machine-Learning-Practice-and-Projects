#### Biary Classification for Lung Cancer
In this excercise we would develop a model that would take values from user and check wether there is chance of cancer to this person or not.

To develop such an model I am using lung_cancer dataset that contains values like Gender, Age, Smooking, CHEST PAIN etc.

> Out dataset is not completely cleaned or maintained so we need to do some data analysis to the dataset, we must convert the dataset into numberic values 


```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
# from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
dataset = pd.read_csv("lung_cancer.csv")
dataset.head()
```

```python
dataset["GENDER"] = dataset["GENDER"].astype('category')
dataset["GENDER"] = dataset["GENDER"].cat.codes
dataset["LUNG_CANCER"] = dataset["LUNG_CANCER"].astype('category')
dataset["LUNG_CANCER"] = dataset["LUNG_CANCER"].cat.codes
dataset.head()
```

```python
dataset.isnull().sum()
```

#### Data is ready
our dataset is cleaned and ready to work as it only contains numeric data for all values, but it contains some unecesorry values. so let's get only required values

```python
dataset = dataset[["GENDER","AGE","SMOKING","ANXIETY","ALLERGY","ALCOHOL CONSUMING","COUGHING","SHORTNESS OF BREATH","CHEST PAIN","LUNG_CANCER"]]
dataset.head()
```

#### Let's Get X and y

```python
X = dataset.drop(columns= "LUNG_CANCER")
y = dataset["LUNG_CANCER"]
```

```python
X.head()
```

```python
X.shape
```

```python
y.head()
```

#### Split data into Training and Testing

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

#### Let's Check our scaled data

Our data would look similer to each other not the too much high or low values. and also it converts the dataset format (series format) to 2D array

#### Lets fit the data to Model now

```python
log_model = LogisticRegression(random_state=0,C=1,fit_intercept=False,max_iter=100)
```

```python
log_model.fit(X_train, y_train)
```

#### Let's check by new data to predict

```python
y_predicton = log_model.predict(X_test)
```

```python
y_predicton
```

#### Let's Check score

```python
log_model.score(X_train, y_train)*100
```

```python
log_model.score(X_test, y_test)*100
```

```python
log_model.score(X_test, y_predicton)*100
```

#### Let's check the probability of Cancer

```python
result = np.round(log_model.predict_proba([[1,45,1,1,1,1,1,1,1]]))
# 0 => No
# 1 => Yes
print(f"Proability of Cancer is : {result}")
```

```python
result = np.round(log_model.predict_proba([[1,15,1,0,0,0,1,1,0]]))
# 0 => No
# 1 => Yes
print(f"Proability of Cancer is : {result[0][1]}")
```

```python
result = np.round(log_model.predict_proba([[0,15,1,1,1,1,1,1,1]]))
# 0 => No
# 1 => Yes
print(f"Proability of Cancer is : {result[0][1]}")
```
