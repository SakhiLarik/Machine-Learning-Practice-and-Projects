```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
import matplotlib.pyplot as plt
```

### Build Data
Lets build some random data for Linear Regression and polynomial Regression

```python
# Linear Regression
#  y = a + bx => a = 4, b = 5 and np.random.randn(100,1) some noise
X = np.random.rand(100,1)
y = 4 + 5 * X + 0.4 * np.random.randn(100,1)
```

##### Visualize data to check if it is linear

```python
plt.scatter(X,y)
```

#### Let's build some data for polynomial regression


```python
# polynomial Regression
# y = a. + bx + cx**2 => a = 2, b = 4, c = 6 and np.random.randn(100,1) some noise
X = 4 * np.random.rand(100,1) - 2
y = 4 + 2 * X + 5 * X**2 + np.random.randn(100,1)
```

```python
plt.scatter(X,y)
```

#### Imoprt Modules
In order to work with polynomial regresion we have to import polynomial features

```python
# we set polynomial features with degree
# degree explains what is the maximium power on X
poly_features = PolynomialFeatures(degree=2)
```

#### Set X_Poly not X for training
for training purpose we have to extract polynomial values from X not exactly the X values and we could use X_poly and y to train the model

in order to convert from X to X Polynomial we have to use method .fit_transform

```python
X_poly = poly_features.fit_transform(X)
```

#### Train or Fit the model

```python
poly_model = LinearRegression().fit(X_poly,y)
```

```python
# Generate new polynomial values for prediction model
X_train = np.linspace(-2,2,100).reshape(-1,1)
X_train_poly = poly_features.transform(X_train)
prediction = poly_model.predict(X_train_poly)
```

#### Check by Visualizing and comparing

```python
plt.scatter(X,y,c='blue')
plt.plot(X_train,prediction,c='red')
```
