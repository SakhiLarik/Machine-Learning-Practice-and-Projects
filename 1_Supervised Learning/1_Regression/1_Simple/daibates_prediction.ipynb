##### Import Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

##### Load Data

```python
diabetes = pd.read_csv("../../Data/diabetes.csv",delimiter='\t')
diabetes.head()
```

##### X and y Setup
I consider age as X and Y as y

```python
X = diabetes[['AGE']]
y = diabetes[['Y']]
```

#### Close look at X and y

```python
X.head()
```

```python
y.head()
```

#### Breakdown X and y for train and test

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

##### Create a model and fit with X_train and y_train

```python
diabetsModel = LinearRegression().fit(X_train,y_train)
```

##### Predict From Model through X_train Data

```python
prediction = diabetsModel.predict(X_test)
prediction
```

#### Visualize to distinguish between actual data and model

```python
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, prediction, color='blue', linewidth=3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```
