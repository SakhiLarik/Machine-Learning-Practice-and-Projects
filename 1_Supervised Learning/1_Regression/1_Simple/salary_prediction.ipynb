##### Import Libraries and Modules Required

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

#### Setup Data

```python
salary = pd.read_csv("../../Data/salary.csv")
salary.head()
```

```python
salary.shape
```

```python
X = salary[["Experience"]]
y = salary[["Salary"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
y_test.head()
```

```python
y_train.head()
```

```python
X_test.head()
```

```python
X_train.head()
```

```python
salaryModel = LinearRegression().fit(X_train,y_train)
```

```python
prediction = salaryModel.predict(X_train)
```

```python
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_train, prediction, color='red', linewidth=1)
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('A Graph Plot Showing Experience and Salary')
plt.show()
```

##### Test and Evaluating (Modal Fitness)

```python
print(f"Training Model Score: {salaryModel.score(X_train,y_train)*100}")
```

```python
print(f"Testing Model Score: {salaryModel.score(X_test,y_test)*100}")
```

```python
print(f"Prediction Score: {salaryModel.score(X_train,prediction)*100}")
```
