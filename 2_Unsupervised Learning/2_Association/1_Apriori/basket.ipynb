#### Data Encoding
Convert data rows into columns. 
1. Each unique item in the description is a column or item that the customer have purchased.
2. Check the rows to find if this item is purchased or not.
3. If this item is purchased set it to 1 else 0

```python
members = basket["Member"].unique()
members
```

```python
set_columns = basket["Description"].unique()
set_columns
```

```python
set_columns[0:5]
```

```python
data = []
for member in members:
    selected_member = basket[basket["Member"] == member]
    for item in set_columns:
        finder = (selected_member[selected_member["Description"] == item])
        if(finder.shape[0]>=1):
            data.append(1)
        if(finder.shape[0]<=0):
            data.append(0)
```

We usually need to do hot encoding in order to set all values in either 0 or 1
* if value of an item >= 1 it is 1
* if value of an item <= 0 it is 0

But our data is already set for this

```python
basket_set = MyBasket.drop(columns=["Members"])
basket_set.head()
```

```python
encoder = TransactionEncoder()
```

```python
encodedData = encoder.fit_transform(basket_set)
```

## Do Machine Learning Now Generate Frequent Itemset

```python
freq_item = apriori(basket_set,min_support=0.5,use_colnames=True)
```

```python
freq_item.head()
```

### Generate and Apply Rules

```python
rules = association_rules(freq_item,metric="lift",min_threshold=1)
```
