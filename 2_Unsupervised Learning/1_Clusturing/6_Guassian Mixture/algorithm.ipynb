### Guassian Mixture

```python
from sklearn.mixture import GaussianMixture
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Data Preparation

```python
data = pd.read_csv("../../data/Mall_Customers.csv")
data = data.drop(columns=["CustomerID","Age","Gender"])
data.head()
```

```python
data["X"] = data["Annual Income (k$)"]
data["Y"] = data["Spending Score (1-100)"]
data = data.drop(columns=["Annual Income (k$)","Spending Score (1-100)"])
data.head()
```

```python
plt.scatter(data["X"],data["Y"])
```

### Machine Learning

```python
gm = GaussianMixture(n_components=6,covariance_type="full")
```

```python
clusters = gm.fit_predict(data[["X"]],data[["Y"]])
np.unique(clusters)
```

```python
data["cluster"] = clusters
```

```python
df1 = data[data['cluster'] == 0]
df2 = data[data['cluster'] == 1]
df3 = data[data['cluster'] == 2]
df4 = data[data['cluster'] == 3]
df5 = data[data['cluster'] == 4]
df6 = data[data['cluster'] == 5]
```

```python
plt.scatter(df1["X"],df1["Y"])
plt.scatter(df2["X"],df2["Y"])
plt.scatter(df3["X"],df3["Y"])
plt.scatter(df4["X"],df4["Y"])
plt.scatter(df5["X"],df5["Y"])
plt.scatter(df6["X"],df6["Y"])
```
