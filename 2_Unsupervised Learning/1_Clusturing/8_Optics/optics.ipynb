### OPTICS
It is the modification of DBSCAN classification

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import OPTICS
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv("../../data/moon.csv")
df.head()
```

```python
plt.scatter(df["X"],df["Y"])
```

```python
op = OPTICS(max_eps=5,min_samples=30,min_cluster_size=2)
```

```python
clusters = op.fit_predict(df[["Y"]])
np.unique(clusters)
```

```python
df["clusters"] = clusters
```

```python
plt.plot(df[df['clusters'] == -1]['X'],df[df['clusters'] == -1]['Y'],'.',c='r')
plt.plot(df[df['clusters'] == 0]['X'],df[df['clusters'] == 0]['Y'],'.',c='b')
plt.plot(df[df['clusters'] == 1]['X'],df[df['clusters'] == 1]['Y'],'.',c='g')
plt.plot(df[df['clusters'] == 2]['X'],df[df['clusters'] == 2]['Y'],'.')
plt.plot(df[df['clusters'] == 3]['X'],df[df['clusters'] == 3]['Y'],'.')
```

```python
df = pd.read_csv("../../data/Mall_Customers.csv")
```

```python
df["X"] = df['Annual Income (k$)']
df["Y"] = df['Spending Score (1-100)']
df.head()
```

```python
op = OPTICS(max_eps=5,min_samples=23,min_cluster_size=5)
```

```python
clusters = op.fit_predict(df[["Y"]])
np.unique(clusters)
```

```python
df['clusters'] = clusters
```

```python
plt.plot(df[df['clusters'] == -1]['X'],df[df['clusters'] == -1]['Y'],'.',c='r')
plt.plot(df[df['clusters'] == 0]['X'],df[df['clusters'] == 0]['Y'],'.')
plt.plot(df[df['clusters'] == 1]['X'],df[df['clusters'] == 1]['Y'],'.')
plt.plot(df[df['clusters'] == 2]['X'],df[df['clusters'] == 2]['Y'],'.')
plt.plot(df[df['clusters'] == 3]['X'],df[df['clusters'] == 3]['Y'],'.')
plt.plot(df[df['clusters'] == 4]['X'],df[df['clusters'] == 4]['Y'],'.')
```
