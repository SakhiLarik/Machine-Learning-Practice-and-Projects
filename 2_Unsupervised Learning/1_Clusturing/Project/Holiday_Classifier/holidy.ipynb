#### Holiday classifier
This project would determine wether it is holidy or working day by looking up the wake up time

> Usually working day wake up time is in between 05 to 10

> Otherwise holiday wake up time is in between 12 to 03

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("../data/holidy.csv")
data.head()
```

```python
KModel = KMeans(n_clusters=2,n_init="auto")
```

```python
KModel.fit(data[["time"]])
```

```python
cluster = KModel.predict(data[["time"]])
cluster
```

```python
data["cluster"] = cluster
```

```python
days = ["working","holiday"]
data["holiday"] = data["cluster"].apply(lambda x: days[x])
```

```python
data.tail()
```

```python
plt.scatter([data["time"]],data["cluster"])
plt.xlabel("Wake up time")
plt.ylabel("Day Type (0 = Holiday, 1 = Working)")
```

```python
# lets assume I am waking up at 4,5 AM and 4,5 PM
KModel.predict([[4],[5],[16],[17]])
# 1, 1, 0, 0 => Working, Working, Holiday, Holiday
```
