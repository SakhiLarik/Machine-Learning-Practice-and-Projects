### Sparse

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.cluster import MeanShift
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import Birch
from sklearn.cluster import OPTICS
```

```python
sparse = pd.read_csv("./colors/sparse.csv")
sparse.head()
```

```python
plt.scatter(sparse[sparse["color"] == 0]["x"],sparse[sparse["color"] == 0]["y"],c='r')
plt.scatter(sparse[sparse["color"] == 1]["x"],sparse[sparse["color"] == 1]["y"],c='b')
plt.scatter(sparse[sparse["color"] == 2]["x"],sparse[sparse["color"] == 2]["y"],c='g')
```

```python
sse = []
for k in range(1,5):
    KM = KMeans(n_clusters=k)
    KM.fit_predict(sparse[["x"]],sparse[["y"]])
    sse.append(KM.inertia_)

plt.plot(range(1,5),sse)
```

```python
MS = MeanShift(bandwidth=30,min_bin_freq=10)
clusters = MS.fit_predict(sparse[["y"]])
```

```python
np.unique(clusters)
```

```python
sparse["cluster"] = clusters
```

```python
plt.scatter(sparse[sparse["cluster"] == 0]["x"],sparse[sparse["cluster"] == 0]["y"])
plt.scatter(sparse[sparse["cluster"] == 1]["x"],sparse[sparse["cluster"] == 1]["y"])
plt.scatter(sparse[sparse["cluster"] == 2]["x"],sparse[sparse["cluster"] == 2]["y"])
```
