### BIRCH
Balanced Iterative Reducing and Clustring using Heirarchies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import Birch
```

```python
df = pd.read_csv("../../data/Mall_Customers.csv")
df["x"] = df["Annual Income (k$)"]
df["y"] = df["Spending Score (1-100)"]
```

```python
df.head()
```

```python
plt.scatter(df["x"],df["y"])
```

```python
birch = Birch(threshold=1,n_clusters=5,branching_factor=100)
```

```python
clusters = birch.fit_predict(df[["y"]])
np.unique(clusters)
```

```python
df['clusters'] = clusters
```

```python
plt.scatter(df[df['clusters'] == 0]['x'],df[df['clusters'] == 0]['y'])
plt.scatter(df[df['clusters'] == 1]['x'],df[df['clusters'] == 1]['y'])
plt.scatter(df[df['clusters'] == 2]['x'],df[df['clusters'] == 2]['y'])
plt.scatter(df[df['clusters'] == 3]['x'],df[df['clusters'] == 3]['y'])
plt.scatter(df[df['clusters'] == 4]['x'],df[df['clusters'] == 4]['y'])
```

```python
moon = pd.read_csv('../../data/moon.csv')
moon.head()
```

```python
plt.scatter(moon["X"],moon["Y"])
```

```python
birch = Birch(n_clusters=2,threshold=0.1,branching_factor=100)
clusters = birch.fit_predict(moon[["Y"]])
np.unique(clusters)
```

```python
moon["cluster"] = clusters
```

```python
plt.scatter(moon[moon['cluster']== 0]["X"],moon[moon['cluster'] == 0]["Y"])
plt.scatter(moon[moon['cluster']== 1]["X"],moon[moon['cluster'] == 1]["Y"])
```
